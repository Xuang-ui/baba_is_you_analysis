```python
# 返回一个 DataFrame: 每行是一个 interaction, index 为 interaction id, 列为 describe_action_sample 返回的 summary 字段
import pandas as pd

def describe_action_sample(gs, interact, post_state_key, action):
    # 计算并返回一个扁平的字典，值为可序列化的标量（number/bool/string）
    info = {}
    next_grid = Gridmap.state_index[post_state_key]

    # 1 greedy (reach more is better)
    try:
        man_before = gs['info']['man_dist'][1]
        man_after = next_grid['info']['man_dist'][1] if next_grid is not None else man_before
        info['man'] = -0.9**man_before + 0.9**man_after
    except Exception:
        info['man'] = 0

    try:
        game_before = gs['info']['game_dist'][1]
        game_after = next_grid['info']['game_dist'][1]
        info['game'] = -0.9**game_before + 0.9**game_after
    except Exception:
        info['game'] = 0

    try:
        bound_before = gs['info']['bound_dist'][1]
        bound_after = next_grid['info']['bound_dist'][1]
        info['bound'] = -0.9**bound_before + 0.9**bound_after
    except Exception:
        info['bound'] = 0
    
    try:
        com_before = gs['info']['com_dist'][1]
        com_after = next_grid['info']['com_dist'][1]
        info['com'] = -0.9**com_before + 0.9**com_after
    except Exception:
        info['com'] = 0

    # 3 curious (more interactions after is better)
    try:
        before_num = gs['inter']['num_inter']
        after_num = next_grid['inter']['num_inter']
        info['inters'] = -before_num + after_num
    except Exception:
        info['inters'] = 0

    # 4 rule first (Jaccard distance between rule lists)
    def jaccard_similarity(a, b):
        interc, unionc = 0, 0
        for a_i, b_i in zip(a, b):
            interc += len(set(a_i) & set(b_i))
            unionc += len(set(a_i) | set(b_i))
        return interc / unionc if unionc > 0 else 1
    try:
        rules_before = gs['info'].get('rules', [])
        rules_after = next_grid['info'].get('rules', []) if next_grid is not None else rules_before
        info['rules'] = 1 - jaccard_similarity(rules_before, rules_after)
    except Exception:
        info['rules'] = 0

    # 5 connectivity change
    try:
        count = 0
        game_before = gs['info'].get('game_dist', [])
        game_after = next_grid['info'].get('game_dist', game_before) if next_grid is not None else game_before
        for d1, d2 in zip(game_before, game_after):
            count += 0.9**d2 - 0.9**d1
        info['games'] = count
    except Exception:
        info['games'] = 0

    # 6 connectivity change
    try:
        count = 0
        game_before = gs['info'].get('bound_dist', [])
        game_after = next_grid['info'].get('bound_dist', game_before) if next_grid is not None else game_before
        for d1, d2 in zip(game_before, game_after):
            count += 0.9**d2 - 0.9**d1
        info['bounds'] = count
    except Exception:
        info['bounds'] = 0

    # 7 connectivity change
    try:
        count = 0
        game_before = gs['info'].get('com_dist', [])
        game_after = next_grid['info'].get('com_dist', game_before) if next_grid is not None else game_before
        for d1, d2 in zip(game_before, game_after):
            count += 0.9**d2 - 0.9**d1
        info['coms'] = count
    except Exception:
        info['coms'] = 0
               

    info['random'] = 0
    # 2 lazy (shorter dist better)
    info['dist'] = -interact.get('trans', {}).get('dist', 0)
    info['repeat'] = 0.9**int(action)

    return info

def describe_action_space(grid):
    gs = grid.summary()
    action_space = gs['inter']['inters']
    rows = []
    for action_sample in action_space:
        interact = Interaction.action_index[action_sample]
        post_space = interact['trans']['post_state']
        for action, post_state in post_space.items():
            summary = describe_action_sample(gs, interact, post_state, action)
            summary['_action_id'] = f'{action_sample}_{action}'
            # print(action, post_state, post_state in Gridmap.state_index)
            rows.append(summary)
    raw = pd.DataFrame(rows).set_index('_action_id')
    ranked = raw.rank(method='average', ascending=False)
    norm = (len(raw) - ranked) / (len(raw) - 1)
    return raw, norm

# 调用示例并展示
df_actions, df_ranked = describe_action_space(grid)
```
